```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('goofy.jpg')
img.shape
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_img.shape
```

```python
#Compute image histogram.

hist = cv2.calcHist([gray_img], [0], None, [256], [0, 256])

#Plot histogram.

plt.plot(hist)
plt.title("Histogram")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(gray_img.ravel(), bins=256, range=[0, 256])
plt.xlabel('pixel intensity')
plt.ylabel('number of pixels')
plt.title('Histogram')
plt.show()
```

```python
#Identify dark and bright regions.

dark_reg = cv2.inRange(gray_img, 0, 50)

bright_reg = cv2.inRange(gray_img, 175, 255)

plt.subplot(1,2,1)
plt.imshow(dark_reg, cmap="gray")
plt.title("Dark Regions")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(bright_reg, cmap="gray")
plt.title("Bright Regions")
plt.axis("off")

plt.show()
```

```python
#Perform histogram equalization.

equalized = cv2.equalizeHist(gray_img)
cv2_imshow(equalized)
```

```python
#Plot equalized histogram.

plt.subplot(1,2,1)
plt.imshow(gray_img, cmap="gray")
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(equalized, cmap="gray")
plt.title("Equalized")
plt.axis("off")

plt.show()
```

```python
#Compare histograms.

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.hist(gray_img.ravel(), bins=256)
plt.title("Original Histogram")

plt.subplot(1,2,2)
plt.hist(equalized.ravel(), bins=256)
plt.title("Equalized Histogram")

plt.show()
```
