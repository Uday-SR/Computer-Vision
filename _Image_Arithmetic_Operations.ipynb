```python
img1 = cv2.imread('goofy.jpg')
img2 = cv2.imread('IMG_20220813_195513.jpg')
```

```python
img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

plt.figure()

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title('Image 1')
plt.axis(False)

plt.subplot(1,2,2)
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
plt.title('Image 2')
plt.axis(False)

plt.show()
```

```python
#Addition

added = cv2.add(img1, img2)
cv2_imshow(added)
```

```python
#Subtraction

subtracted = cv2.subtract(img1, img2)
cv2_imshow(subtracted)
```

```python
#Blend

alpha = 0.7
beta = 0.3

blended = cv2.addWeighted(img1, alpha, img2, beta, 0)
cv2_imshow(blended)
```

```python
#Multiplied

multiplied = cv2.convertScaleAbs(img1, alpha=1.5, beta=0)
cv2_imshow(multiplied)
```

```python
#Save

cv2.imwrite("added.jpg", added)
cv2.imwrite("subtracted.jpg", subtracted)
cv2.imwrite("blended.jpg", blended)
cv2.imwrite("multiplied.jpg", multiplied)
```

```python
#Compare

titles = [
    "Original Image 1",
    "Original Image 2",
    "Added",
    "Subtracted",
    "Blended",
    "Intensity Multiplied"
]

images = [
    cv2.cvtColor(img1, cv2.COLOR_BGR2RGB),
    cv2.cvtColor(img2, cv2.COLOR_BGR2RGB),
    cv2.cvtColor(added, cv2.COLOR_BGR2RGB),
    cv2.cvtColor(subtracted, cv2.COLOR_BGR2RGB),
    cv2.cvtColor(blended, cv2.COLOR_BGR2RGB),
    cv2.cvtColor(multiplied, cv2.COLOR_BGR2RGB)
]

plt.figure(figsize=(15,10))

for i in range(len(images)):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
#Pixel changes

row = 100
col = 100

print("Image1 Pixel:", img1[row, col])
print("Image2 Pixel:", img2[row, col])

print("Added:", added[row, col])
print("Subtracted:", subtracted[row, col])
print("Blended:", blended[row, col])
print("Multiplied:", multiplied[row, col])
```
