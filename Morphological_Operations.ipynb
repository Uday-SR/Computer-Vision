```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('goofy.jpg')
img = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)[1]
cv2_imshow(img)
```

```python
kernel = np.ones((3,3), np.uint8)
```

```python
erosion = cv2.erode(img, kernel, iterations=1)
cv2_imshow(erosion)
```

```python
dilation = cv2.dilate(img, kernel, iterations=1)
cv2_imshow(dilation)
```

```python
opening = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
cv2_imshow(opening)
```

```python
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
cv2_imshow(closing)
```

```python
plt.figure(figsize=(8,5))

plt.subplot(1,5,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_RGB2BGR), cmap="gray")
plt.title("Original")
plt.axis("off")

plt.subplot(1,5,2)
plt.imshow(cv2.cvtColor(erosion, cv2.COLOR_RGB2BGR), cmap="gray")
plt.title("Erosion")
plt.axis("off")

plt.subplot(1,5,3)
plt.imshow(cv2.cvtColor(dilation, cv2.COLOR_RGB2BGR), cmap="gray")
plt.title("Dilation")
plt.axis("off")

plt.subplot(1,5,4)
plt.imshow(cv2.cvtColor(opening, cv2.COLOR_RGB2BGR), cmap="gray")
plt.title("Opening")
plt.axis("off")

plt.subplot(1,5,5)
plt.imshow(cv2.cvtColor(closing, cv2.COLOR_RGB2BGR), cmap="gray")
plt.title("Closing")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
#3x3 kernel

kernel3 = np.ones((3,3), np.uint8)
erosion3 = cv2.erode(img, kernel3)

#5x5 kernel

kernel5 = np.ones((5,5), np.uint8)
erosion5 = cv2.erode(img, kernel3)

#7x7 kernel

kernel7 = np.ones((7,7), np.uint8)
erosion7 = cv2.erode(img, kernel3)

#compare

plt.figure(figsize=(8,5))

plt.subplot(1,4,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_RGB2BGR), cmap="gray")
plt.title("Original")
plt.axis("off")

plt.subplot(1,4,2)
plt.imshow(cv2.cvtColor(erosion3, cv2.COLOR_RGB2BGR), cmap="gray")
plt.title("3×3")
plt.axis("off")

plt.subplot(1,4,3)
plt.imshow(cv2.cvtColor(erosion5, cv2.COLOR_RGB2BGR), cmap="gray")
plt.title("5×5")
plt.axis("off")

plt.subplot(1,4,4)
plt.imshow(cv2.cvtColor(erosion7, cv2.COLOR_RGB2BGR), cmap="gray")
plt.title("7×7")
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
print("Original :", np.count_nonzero(img))

print("Erosion :", np.count_nonzero(erosion))

print("Dilation :", np.count_nonzero(dilation))

print("Opening :", np.count_nonzero(opening))

print("Closing :", np.count_nonzero(closing))
```
