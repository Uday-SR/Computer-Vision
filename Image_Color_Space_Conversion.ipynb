```python
img = cv2.imread('goofy.jpg')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
#Convert image to HSV color space.

hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2_imshow(hsv)
```

```python
#Convert image to LAB color space.

lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
cv2_imshow(lab)
```

```python
#Compare different color spaces.

#shapes

print('SHAPES')
print("\nRGB Shape :", img.shape)
print("Gray Shape:", gray.shape)
print("HSV Shape :", hsv.shape)

#pixels values

y, x = 100, 100

print("\nPIXEL VALUES")
print("\nRGB :", img[y, x])
print("Gray:", gray[y, x])
print("HSV :", hsv[y, x])

#individual channels

print('INDIVIDUAL CHANNELS')

print('\nRGB channel')
print(f'\nr = {img[:,:,0]}')
print(f'g = {img[:,:,1]}')
print(f'b = {img[:,:,2]}')

print('\nHSV channel')
print(f'\nh = {hsv[:,:,0]}')
print(f's = {hsv[:,:,1]}')
print(f'v = {hsv[:,:,2]}')
```
