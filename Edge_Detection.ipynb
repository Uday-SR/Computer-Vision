```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('goofy.jpg')
cv2_imshow(img)
```

```python
#Sobel operator

#for horizontal edges

sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)

#for vertical edges

sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

#combine x and y

sobel = cv2.magnitude(sobel_x, sobel_y)

sobel = cv2.convertScaleAbs(sobel)

cv2_imshow(sobel)
```

```python
#Laplacian Operator

laplace = cv2.Laplacian(img, cv2.CV_64F)
laplace = cv2.convertScaleAbs(laplace)
cv2_imshow(laplace)
```

```python
#Canny Edge detector

canny = cv2.Canny(img, 100, 200)
canny = cv2.convertScaleAbs(canny)
cv2_imshow(canny)
```

```python
#compare

plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB), cmap="gray")
plt.title("Original")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(cv2.cvtColor(sobel, cv2.COLOR_BGR2RGB), cmap="gray")
plt.title("Sobel")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(laplace, cmap="gray")
plt.title("Laplacian")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(cv2.cvtColor(canny, cv2.COLOR_BGR2RGB), cmap="gray")
plt.title("Canny")
plt.axis("off")

plt.show()
```

```python
#count edge pixels

sobel_edge_pix = np.count_nonzero(sobel)
laplace_edge_pix = np.count_nonzero(laplace)
canny_edge_pix = np.count_nonzero(canny)

print('Edge pixels (Sobel) : ', sobel_edge_pix)
print('Edge pixels (laplace) : ', laplace_edge_pix)
print('Edge pixels (canny) : ', canny_edge_pix)

#Edge density

sobel_edge_density = np.count_nonzero(sobel)/sobel.size
laplace_edge_density = np.count_nonzero(laplace)/laplace.size
canny_edge_density = np.count_nonzero(canny)/canny.size

print('\nEdge density (Sobel) : ', sobel_edge_density)
print('Edge density (laplace) : ', laplace_edge_density)
print('Edge density (canny) : ', canny_edge_density)

#Sharpness

sob_sharp = sobel.var()
lap_sharp = laplace.var()
can_sharp = canny.var()

print('\nSharpness (Sobel)', sob_sharp)
print('Sharpness (Laplace)', lap_sharp)
print('Sharpness (Canny)', can_sharp)
```
