```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import time
from google.colab.patches import cv2_imshow
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
img = cv2.imread('goofy.jpg')
cv2_imshow(img)
```

```python
#Average Filter

start_avg = time.time()

average = cv2.blur(img, (5, 5))
cv2_imshow(average)

end_avg = time.time()
```

```python
#Gaussian Filter

start_gauss = time.time()

gaussian = cv2.GaussianBlur(img, (5, 5), 0)
cv2_imshow(gaussian)

end_gauss = time.time()
```

```python
#Median Filter

start_med = time.time()

median = cv2.medianBlur(img, 5)
cv2_imshow(median)

end_med = time.time()
```

```python
#Compare results

plt.figure(figsize=(12,8))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(average, cv2.COLOR_BGR2RGB))
plt.title('Average Filter')
plt.axis(False)

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(gaussian, cv2.COLOR_BGR2RGB))
plt.title('Gaussian Filter')
plt.axis(False)

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(median, cv2.COLOR_BGR2RGB))
plt.title('Median Filter')
plt.axis(False)

plt.show()
```

```python
#Standard Deviation

print('Original : ', np.std(img))
print('Average Filter : ', np.std(average))
print('Gaussian Filter : ', np.std(gaussian))
print('Median Filter : ', np.std(median))
```

```python
#Compare filter execution time

print("Exec time for average filter : ", end_avg-start_avg)
print("Exec time for Guassian filter : ", end_gauss-start_gauss)
print("Exec time for Median filter : ", end_med-start_med)
```

```python
#Save

cv2.imwrite('avg.jpg', average)
cv2.imwrite('gauss.jpg', gaussian)
cv2.imwrite('med.jpg', median)
```
