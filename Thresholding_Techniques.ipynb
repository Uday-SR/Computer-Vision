```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('goofy.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img)
```

```python
binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)[1]
cv2_imshow(binary)
```

```python
adaptive = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
cv2_imshow(adaptive)
```

```python
otsu = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
cv2_imshow(otsu)
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,4,1)
plt.imshow(img,cmap="gray")
plt.title("Original")
plt.axis("off")

plt.subplot(1,4,2)
plt.imshow(binary,cmap="gray")
plt.title("Binary")
plt.axis("off")

plt.subplot(1,4,3)
plt.imshow(adaptive,cmap="gray")
plt.title("Adaptive")
plt.axis("off")

plt.subplot(1,4,4)
plt.imshow(otsu,cmap="gray")
plt.title("Otsu")
plt.axis("off")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.subplot(1,3,1)
plt.hist(binary.ravel(), bins=256, range=[0,256])
plt.title("Binary Hist")

plt.subplot(1,3,2)
plt.hist(adaptive.ravel(), bins=256, range=[0,256])
plt.title("Adaptive Hist")

plt.subplot(1,3,3)
plt.hist(otsu.ravel(), bins=256, range=[0,256])
plt.title("Otsu Hist")

plt.tight_layout()
plt.show()
```

```python
cv2.imwrite('binary.jpg', binary)
cv2.imwrite('Adaptive.jpg', adaptive)
cv2.imwrite('otsu.jpg', otsu)
```
